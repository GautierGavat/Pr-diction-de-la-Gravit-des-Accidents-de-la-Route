```python
def clean_onisr_data(df):
    df = df.replace(-1, np.nan)
    cols_zero_is_nan = ['trajet', 'locp', 'actp', 'lat', 'long']
    df[cols_zero_is_nan] = df[cols_zero_is_nan].replace(0, np.nan)

    nan_ratio = df.isna().mean()
    cols_to_drop = nan_ratio[nan_ratio > 0.9].index
    print(f"Colonnes supprimées (>90% de valeurs nulles) : {list(cols_to_drop)}")
    df = df.drop(columns=cols_to_drop)

    df['vma'] = df['vma'].fillna(df.groupby('catr')['vma'].transform('median'))

    df['date_dt'] = pd.to_datetime(df['Date'])
    df['age_usager'] = df['date_dt'].dt.year - df['an_nais']
    
    critical_cols = [c for c in ['atm', 'lum', 'int', 'grav'] if c in df.columns]
    df = df.dropna(subset=critical_cols)
    
    #Suppression des colonnes qui ont plus de 50% de valeurs nulles
    
    limit = len(df) * 0.5
    df = df.dropna(thresh=int(limit), axis=1)

    return df

df_clean = clean_onisr_data(df)
```

secu1
Le renseignement du caractère indique la présence et l’utilisation de l’équipement de sécurité :
-1 – Non renseigné
0 – Aucun équipement
1 – Ceinture
2 – Casque
12
3 – Dispositif enfants
4 – Gilet réfléchissant
5 – Airbag (2RM/3RM)
6 – Gants (2RM/3RM)
7 – Gants + Airbag (2RM/3RM)
8 – Non déterminable
9 – Autre

```python
def nouvelle_valeur(valeur):
    protege = [1, 2, 3, 4, 5, 6, 7]
    non_protege = [0]
    if valeur in protege:
            return 1
    elif valeur in non_protege:
            return 2
    else :
            return 3
            
df_clean['protection'] = df_clean['secu1'].apply(nouvelle_valeur)
df_clean = df_clean.drop(columns='secu1')
df_clean
        
```
