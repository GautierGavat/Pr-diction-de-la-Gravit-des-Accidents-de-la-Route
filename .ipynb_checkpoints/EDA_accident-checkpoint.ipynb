```python
df_bis = df_2024.drop("Num_Acc", axis=1)
df_bis = df_bis.drop("hrmn", axis=1)
df_bis = df_bis.drop("dep", axis=1)
df_bis = df_bis.drop("com", axis=1)
df_bis = df_bis.drop("id_usager", axis=1)
df_bis = df_bis.drop("id_vehicule", axis=1)
df_bis = df_bis.drop("num_veh", axis=1)
```

```python
df_bis
```

```python
df_bis["nbv"] = df_bis["nbv"].astype(int)
```

```python
df_bis["nbv"].unique()
```

```python
(df_bis["nbv"] == "#VALEURMULTI").sum()
```

```python
df_bis["nbv"] = pd.to_numeric(df_bis["nbv"], errors="coerce")
df_bis.loc[df_bis["nbv"] > 1, "nbv"].mean()
df_bis["nbv"] = df_bis["nbv"].fillna(df_bis.loc[df_bis["nbv"] > 1, "nbv"].mean()).round().astype(int)
```

```python
df_bis["pr"] = df_bis["pr"].astype(int)
```

```python
df_bis[df_bis["pr"].astype(str).str.contains("\xa0", na=False)]
```

```python
df_bis["pr"] = (
    df_bis["pr"]
    .astype(str)
    .str.replace("\xa0", "", regex=False)  # espace insécable
    .str.replace(" ", "", regex=False)     # espace normal
    .astype(int)
)
```

```python
df_bis["pr1"] = df_bis["pr1"].astype(int)
df_bis["larrout"] = df_bis["larrout"].astype(float)
```

```python
df_bis["pr1"] = (
    pd.to_numeric(
        df_bis["pr1"]
        .astype(str)
        .str.replace(r"\s+", "", regex=True),
        errors="coerce"
    )
    .astype("Int64")
)
```

```python
df_bis["larrout"] = (
    df_bis["larrout"]
    .astype(str)
    .str.replace(",", ".", regex=False)
    .astype(float)
)
```

```python
df_bis["lat"] = df_bis["lat"].astype(float)
df_bis["long"] = df_bis["long"].astype(float)
```
